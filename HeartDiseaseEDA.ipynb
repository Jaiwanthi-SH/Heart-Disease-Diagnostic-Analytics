Heart Disease variation based on blood vesssel distribution

```python
plt.hist(df[df['target'] == 0]['ca'], alpha=1.0, label='No Heart Disease')
plt.hist(df[df['target'] == 1]['ca'], alpha=1.0, label='Heart Disease')
plt.title('Blood Vessel distribution VS status')
_ = plt.legend()
```
